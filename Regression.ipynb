### 회귀법

##### 종속변수와 독립변수간의 관계를 알아보기 위한 방법으로 다음과 같은 종류가 있다.
-  1) 선형 회귀 Linear-Regression
-  2) 비선형 회귀 Non-Lenear-Regression
-  3) 다항식 회귀 Polynomial

- 선형 회귀식
> $y = \beta_{0} + \beta_{1}x$ .   
> $e = y - (\beta_{0} + \beta_{1}x) = y - \hat{y}$ .     

이러한 error를 측정할 때 최소자승법을 통해 측정한다.    
error 즉, 오차의 합이 최소가 되는 선형 회귀식을 구하는 것이 선형 회귀법의 원리이다.    

- 최소자승법 Least-squared Method
> $\epsilon = \displaystyle\sum_{i=1}^{n}\epsilon_{i}^{2} = \displaystyle\sum_{i=1}^{n}[y_i - (\beta_0 + \beta_1x_i)]^2$  

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

# x = np.array([1,2,3,4,5])
y = np.array([8.5, 7.7, 6.9, 4.5, 2])

# beta0_nom = np.sum(y) * np.sum(x**2) - np.sum(x) * np.sum(x*y)
# denomk = len(x) * np.sum(x**2) - (np.sum(x)) ** 2
# beta0 = beta0_nom / denom

# beta1_nom = len(x) * np.sum(x*y) - np.sum(x) * np.sum(y)
# beta1 = beta1_nom /denom
# print(beta1_nom)
# print(beta1)   

XT = np.array([[1,1,1,1,1], [1,2,3,4,5], [1, 4, 9, 16, 25]])
X = XT.T
beta = np.linalg.inv(XT @ X) @ (XT @ y) # 결과 값 beta 는 2x1 형태를 띄게 됨
beta0, beta1, beta2 = beta[0], beta[1], beta[2]

print(beta0, beta1, beta2) 
# yhat = beta0 + beta1 * x
yhat = X @ beta
print(yhat)

# R2 Score
ymean = np.mean(yhat)
up = 0
bottom = 0
for idx, yi in enumerate(y):
    up += pow(2, yi - yhat[idx])
    bottom += pow(2, yi - ymean)

acc = 1 - up/bottom
print(f'R2 score : {acc}')

plt.scatter(XT[1], y)
plt.plot(XT[1], beta0 * XT[0] + beta1 * XT[1] + beta2 * XT[2], color = "red") 
plt.xlabel("Cups of Coffee")
plt.ylabel("Sleep time")
plt.show()
```
