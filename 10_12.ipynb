```python
import numpy as np

arr = np.arange(1, 10, 0.1)
sine = []

for i in arr:
    sine.append(np.sin(i))

```

```python
import matplotlib.pyplot as plt

plt.plot(sine)
plt.show()
```

```python
A = np.array([[1,2,3], [2,3,4], [3,4,5]])
print(A)

hap = 0
for i in range(3):
    for j in range(3):
        hap += A[i,j]
        
print(f'hap = {hap}')
```
